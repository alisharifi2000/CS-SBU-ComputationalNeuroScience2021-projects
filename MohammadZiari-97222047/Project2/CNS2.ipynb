```python
from brian2 import *
import random
%matplotlib inline
```

```python
eq = ''' dv/dt = (I-v)/tau : 1
I : 1
tau : second '''
```

# **ONE**

## **Group1**

```python
start_scope()

I = []
for i in range(1000):
  I.append(random.random())

Pop = NeuronGroup(1000, eq, threshold='v>-50', reset='v=-65' , method='exact')
Pop.I = I
Pop.tau = 1000*ms
Pop_E = Pop[:800]
Pop_I = Pop[800:]

Syn_E = Synapses(Pop_E, Pop, on_pre='v_post += 0.05')
Syn_I = Synapses(Pop_I, Pop, on_pre='v_post += 0.1')

Syn_E.connect(p=0.05)
Syn_I.connect(p=0.10)

Mon = StateMonitor(Pop, 'v', record=True)
Mon_E = SpikeMonitor(Pop[:800])
Mon_I = SpikeMonitor(Pop[800:])

run(1000*ms)
```

```python
# I - T
plot(I , label=i+1)
xlabel('Time')
ylabel('I')
legend();
```

```python
# Raster_Plot
plt.scatter(Mon_E.t/ms, Mon_E.i, label='Excitatory')
plt.scatter(Mon_I.t/ms, Mon_I.i+800, label='Inhibitory')
legend()
```

```python
# V - T
j=0 
i=0
for i in range(2):
  plot(Mon.t/ms, Mon.v[i+j], label=i+1)
  j = 800
xlabel('Time')
ylabel('Membrene Potiential')
legend();
```

# **TWO**

## **Group1**

```python
start_scope()

I1 = []
I2 = []
I3 = []
for i in range (200):
  I1.append(random.random())
  I2.append(random.random())
for i in range(100):
  I3.append(random.random())

Pop_E1 = NeuronGroup(200, eq, threshold='v>-50', reset='v = -65')
Pop_E2 = NeuronGroup(200, eq, threshold='v>-50', reset='v = -65')
Pop_I = NeuronGroup(100, eq, threshold='v>-50', reset='v = -65')

Pop_E1.I = I1
Pop_E2.I = I2
Pop_I.I = I3
Pop_E1.tau = 1000*ms
Pop_E2.tau = 1000*ms
Pop_I.tau = 1000*ms

Syn_1 = Synapses(Pop_E1, Pop_E1, on_pre='v_post += 0.3')
Syn_2 = Synapses(Pop_E2, Pop_E2, on_pre='v_post += 0.1')
Syn_3 = Synapses(Pop_E1, Pop_I, on_pre='v_post += 0.05')
Syn_4 = Synapses(Pop_E2, Pop_I, on_pre='v_post += 0.2')
Syn_5 = Synapses(Pop_I, Pop_I, on_pre='v_post -= 0.4')

Syn_1.connect(p=0.1)
Syn_2.connect(p=0.05)
Syn_3.connect(p=0.1)
Syn_4.connect(p=0.2)
Syn_5.connect(p=0.1)

Spike_M1 = SpikeMonitor(Pop_E1)
Spike_M2 = SpikeMonitor(Pop_E2)
Spike_M3 = SpikeMonitor(Pop_I)

M1 = StateMonitor(Pop_E1, 'v', record=True)
M2 = StateMonitor(Pop_E2, 'v', record=True)
M3 = StateMonitor(Pop_I, 'v', record=True)

run(1000*ms)
```

```python
# I - T (exc1)
plot(I1 , label=i+1)
xlabel('Time')
ylabel('I_Exc1')
legend();
```

```python
# I - T (exc2)
plot(I2 , label=i+1)
xlabel('Time')
ylabel('I_Exc2')
```

```python
# I - T (inh)
plot(I3 , label=i+1)
xlabel('Time')
ylabel('I_Exc3')
```

```python
# Raster_Plot
plt.scatter(Spike_M1.t/ms, Spike_M1.i, label='Excitatory1')
plt.scatter(Spike_M2.t/ms, Spike_M2.i, label='Excitatory2')
plt.scatter(Spike_M3.t/ms, Spike_M3.i, label='Inhibitory')
legend()
```

```python
# V - T
plot(M1.t/ms, M1.v[i], label=1)
plot(M2.t/ms, M2.v[i], label=2)
plot(M3.t/ms, M3.v[i], label=3)
xlabel('Time')
ylabel('Membrene Potiential')
legend(); 
```
